```python
#Connecting to hospital_db

import urllib
import pandas as pd
from sqlalchemy import create_engine

# 1. Define your connection parameters
server = 'localhost'
database = 'hospital_db'

# 2. Build the standard connection string
conn_str = (
    f"Driver={{ODBC Driver 17 for SQL Server}};"
    f"Server={server};"
    f"Database={database};"
    f"Trusted_Connection=yes;"
)

# 3. URL-encode the connection string for SQLAlchemy compatibility
params = urllib.parse.quote_plus(conn_str)
engine_url = f"mssql+pyodbc:///?odbc_connect={params}"

try:
    # 4. Create the SQLAlchemy engine
    engine = create_engine(engine_url)
    print("SQLAlchemy engine created successfully!")

    # 5. Quick Test: Load your encounter date data into a Pandas DataFrame
    query = "SELECT TOP 5 * FROM gold.dim_encounter_date"
    df = pd.read_sql(query, con=engine)
    
    # Display the results
    print("\nConnection verified! Previewing data:")
    print(df)

except Exception as e:
    print(f"Database connection failed: {e}")
```
